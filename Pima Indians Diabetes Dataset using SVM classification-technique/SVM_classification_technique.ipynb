# Pima indians diabetes dataset analysis using SVM Classification Technique 

```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
```

```python
df
```

```python
df.head(20)
```

```python
#Check the weather any null value present here or not  
df.isnull().sum()
```

```python
#Drop the Label value from this data_set
x = df.drop(['Class'],axis =1 )
```

```python
x
```

```python
# Separated Label Value 
y = df['Class']
```

```python
y
```

# Train Data

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .20,random_state = 1)
```

```python
xtrain
```

```python
model = SVC()
```

```python
model.fit(xtrain,ytrain)
```

# Test Score

```python
model.score(xtest,ytest)
```

```python
# This model gave 78% acurate answer by using SVM CLassification Technique.
```
